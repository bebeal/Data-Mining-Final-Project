```python
# Entire Dataset
path = 'spotify_dataset.csv'
data = pd.DataFrame(pd.read_csv(path))
data.head()

def cat_onehot(feat_name):
    cat_feats = data[feat_name].tolist()
    genres = set()
    for feat_list in cat_feats:
        temp = feat_list.replace("'", "")
        temp = temp.replace("[", "")
        temp = temp.replace("]", "")
        temp = temp.replace(" ", "")
        feats  = temp.split(",")
        for feat in feats:
            if "pop" in feat:
                genres.add("pop")
            elif "rock" in feat:
                genres.add("rock")
            elif "rap" in feat:
                genres.add("rap")
            elif "hiphop" in feat:
                genres.add("hiphop")
            else:
                genres.add(feat)
    print(genres)
    print(len(genres))

cat_onehot("Genre")
```
